#### Exercise 01. What is the most expensive house in the entire dataset? (★☆☆)

Print the address and price of the selected house. For example:

`The house located at Calle del Prado, Nº20 is the most expensive, and its price is 5000000 USD.`

```python
# TODO
df_sample = pd.read_csv('assets/real_estate.csv', nrows=1000, sep=';')


print(df_sample.columns)

# Ensure "price" is numeric (in case of formatting issues)
df_sample["price"] = pd.to_numeric(df_sample["price"], errors="coerce")

# Find the most expensive house
most_expensive = df_sample.loc[df_sample["price"].idxmax()]

# Display the result
print(f'The house located at {most_expensive["address"]} is the most expensive, and its price is {most_expensive["price"]} USD.')
```

#### Exercise 02. What is the cheapest house in the dataset? (★☆☆)

```python
# TODO
cheapest_house = df_sample.loc[df_sample["price"].idxmin()]

print(f'The house located at {cheapest_house["address"]} is the cheapest, and its price is {cheapest_house["price"]} USD.')
```

#### Exercise 07. What is the average price in the population of "Arroyomolinos (Madrid)"? (★★☆)

Print the value obtained from the 'level5' column.

```python
# TODO

filtered_df = df_sample[df_sample["level5"] == "Arroyomolinos (Madrid)"]
average_price = filtered_df["price"].mean()
print(f'The average price of houses in Arroyomolinos (Madrid) is {average_price:.2f} USD.')



```

#### Exercise 08. Plot the histogram of prices for the population of "Arroyomolinos (Madrid)" and explain what you observe (★★☆)

Print the histogram of the prices and write a brief analysis of the plot in the Markdown cell.

```python
import matplotlib.pyplot as plt

filtered_df = df_sample[df_sample["level5"] == "Arroyomolinos (Madrid)"]

filtered_df["price"] = pd.to_numeric(filtered_df["price"], errors="coerce")

filtered_df = filtered_df.dropna(subset=["price"])

plt.figure(figsize=(10, 6))
plt.hist(filtered_df["price"], bins=20, edgecolor="black", alpha=0.7)

plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.title("Distribution of House Prices in Arroyomolinos (Madrid)")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```
